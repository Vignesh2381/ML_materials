# Visualization

### Histogram: 
1. A histogram is a bar chart that displays the frequency distribution of continuous or discrete data. 
2. It divides the range of the data into intervals (bins) and shows how many data points fall into each bin. 
3. This helps you see the shape of the distribution

### Kernel Density Plot: 
1. Similar to a histogram, a kernel density plot represents the distribution of data points as a smooth curve. 
2. It provides a continuous representation of the data's density

### Box Plot (Box-and-Whisker Plot): 
1. A box plot provides a visual summary of the distribution's median, quartiles, and outliers. 
2. It can show you whether the data is symmetric or skewed and if there are any extreme values

### Empirical Cumulative Distribution Function (ECDF) Plot: 
An ECDF plot shows the cumulative distribution of the data. It helps you understand what proportion of data falls below a certain value

### Histogram with Overlaid Kernel Density: 
Combining a histogram with a kernel density plot can provide a comprehensive view of the distribution, showing both discrete bin counts and continuous density

# Continuous Numeric Data:

### Histogram: 
To understand the distribution of values, identify modes, skewness, and outliers.

### Kernel Density Plot:
Similar to histograms, but provides a smooth density curve.

### Box Plot: 
To visualize the summary statistics, spread, and identify outliers.

# Categorical Data:

### Bar Plot: 
To visualize the frequency or count of each category.

### Pie Chart: 
To show the proportion of each category within a whole (use sparingly, as they can be less effective than other options).
### Count Plot: 
A specialized bar plot for counting occurrences of each category.

# Time Series Data:

### Line Plot: 
To show trends and patterns over time.
### Area Plot: 
Similar to line plot, but the area below the line is filled, highlighting the cumulative effect

# Comparison:

### Side-by-Side Bar Plot: 
To compare frequency or counts across different categories.
### Grouped Box Plot: 
To compare summary statistics of different groups.
# Distribution Comparison:

### Overlayed Histogram/KDE Plot: 
To compare the distributions of two or more groups.

# Categorical and Numerical Data (Distribution of Numerical Data within Categories):

### Box Plot or Violin Plot: 
Use these plots to visualize how the numerical data is distributed within different categories. Each box or violin represents the distribution of the numerical data for a specific category.

### Bar Plot or Swarm Plot: 
While bar plots can show aggregated statistics like mean or median, swarm plots show individual data points, giving you a sense of the data's distribution within each category.

# Categorical and Categorical Data (Distribution of Counts/Frequencies):

### Bar Plot: 
Use a bar plot to visualize the count or frequency of each category combination. Stacked bar plots can give insights into proportions as well.

### Heatmap: 
Create a heatmap where the rows and columns represent different categorical variables, and the cells show the frequency or count of their combinations.

# Numerical and Numerical Data (Distribution of Both Variables):

### Histograms: 
Create histograms for both numerical variables to visualize their individual distributions. You can use different colors or transparency to distinguish the two variables.

### Kernel Density Plot: 
Similar to histograms, kernel density plots can give a smooth representation of the distribution of each numerical variable.

### Marginal Distribution Plots: 
In a scatter plot, you can add marginal distribution plots (histograms or density plots) on the sides to show the individual distributions of both variables

# Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('titanic.csv')
df.dropna(inplace=True)
df
```

# Continuous Data

```python
# Histogram with Overlaid Kernel Density
data=df.copy()
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='blue', label='Age')
plt.xlabel('Numeric Value')
plt.ylabel('Frequency')
plt.title('Histogram with Overlaid Kernel Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Categorical Data

```python
# Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=data, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count Plot')
plt.tight_layout()
plt.show()
```

# Categorical and Numerical Data

```python
sns.barplot(x='Embarked', y='Fare', data=data, ci=None)
```

```python
# Side-by-Side Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Embarked', y='Fare', data=data, ci=None, color='blue', label='Fare')
sns.barplot(x='Embarked', y='Age', data=data, ci=None, color='orange', label='Age')
plt.xlabel('Category')
plt.ylabel('Numeric Value')
plt.title('Side-by-Side Bar Plot')
plt.legend()
plt.tight_layout()
plt.show()
```

# Skewed Data

### Resampling: 
1. Resampling involves creating new samples by duplicating existing ones (oversampling) or removing some instances (undersampling). 
2. This can help balance the class distribution. However, it may lead to overfitting in some cases, so be cautious and evaluate its impact carefully

### Ensemble methods: 
Ensemble methods like Random Forest, Gradient Boosting, or AdaBoost can handle imbalanced data well by combining the results of multiple weak learners

### Feature engineering: 
Crafting informative features can improve the performance of models on imbalanced data. Try to include relevant features that can help distinguish between different classes more effectively

### Data collection: 
If possible, consider collecting more data for the minority class to balance the distribution. Having a more balanced dataset can significantly improve the model's performance

# Transform skewed data 

### Logarithmic transformation: 
Applying the logarithm function to the data can help reduce the impact of extreme values and compress the range, making the distribution more symmetrical.

### Square root transformation: 
Taking the square root of the data can also mitigate the effect of extreme values and reduce skewness.

### Box-Cox transformation: 
The Box-Cox transformation is a family of power transformations that can optimize the data's normality. It includes logarithmic and square root transformations as special cases.

### Inverse transformation: 
The inverse of data (1/x) can be used when the data is exponentially distributed.

### Z-score normalization: 
This method standardizes the data by subtracting the mean and dividing by the standard deviation. It doesn't change the shape of the distribution but centers it at zero and scales it.
